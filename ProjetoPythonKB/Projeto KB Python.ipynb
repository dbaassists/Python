# 03.5.2 - Criar uma Função (Completa) padrão para Importar um Arquivo CSV usando arquivo de parâmetro

```python
paramJson = "..\ProjetoPythonKB\\config\config.json"
file = open(paramJson)
dfJson = json.load(file)

def montaCampos_COMPLETA(colunas,dtype,usecols,parse_dates):
        
    vcolunas = []
    vusecols = []
    vdtype = [] 
    vparse_dates = []
    vcolunasDataType = {}    

    for indlistaCampos in colunas.split(sep=','):
        if indlistaCampos != ',':        
            vcolunas.append(indlistaCampos.strip())   
            
    for indlistausecols in usecols.split(sep=','):
        if indlistausecols != ',':        
            vusecols.append(int(indlistausecols.strip()))

    for indlistadtype in dtype.split(sep=','):
        if indlistadtype != ',':        
            vdtype.append(indlistadtype.strip())
            
    for indlistaparsedates in parse_dates.split(sep=','):
        if indlistaparsedates != ',':        
            vparse_dates.append(int(indlistaparsedates.strip()))

    for itemChave in range(len(vcolunas)):
        vcolunasDataType[vcolunas[itemChave]] = vdtype[itemChave]

    return vcolunas, vcolunasDataType, vusecols, vparse_dates

colunas = dfJson['ConfigArquivo']['colunas']
dtype = dfJson['ConfigArquivo']['dtype']
usecols = dfJson['ConfigArquivo']['usecols']
sep = dfJson['ConfigArquivo']['sep']
encoding = dfJson['ConfigArquivo']['encoding']
header = dfJson['ConfigArquivo']['header']
parse_dates = dfJson['ConfigArquivo']['parse_dates']

vcolunas,vcolunasDataType,vusecols,vparse_dates = montaCampos_COMPLETA(colunas,dtype,usecols,parse_dates)

arquivoCSV = '..\ProjetoPythonKB\Dados\Candidato\consulta_cand_2022_BRASIL.csv'  

df = pd.read_csv(arquivoCSV
                                ,sep=sep
                                ,encoding=encoding
                                ,usecols=vusecols
                                ,names=vcolunas
                                ,dtype=vcolunasDataType
                                ,header=header
                                ,parse_dates=vparse_dates)


df.info()
```

# 08 - Ordenando um DataFrame (sort_values())
* Detalhe: Método sort_values() realiza a odernação de um DataFrame baseado nas colunas.


* No exemplo abaixo, estamos fazendo as seguintes operações:
** Importação de um arquivo csv usando o método Pandas read_csv() onde definimos o delimitador de colunas, o encoding de dados, as colunas, o nome e datatype que cada coluna irá receber;


* Os parâmetros usados no exemplo abaixo foram:
** by='' - Define qual coluna deverá ser ordenada. Caso tenha mais de uma coluna na clausula de ordenação, elas devem vim dentro de conchetes ([]);
** inplace=True - significa que a ordenação deve ser persistida no DataFrame carregado;
** ascending=True - ordenação deverá ser feita do menor pro maior ou conforme o alfabeto [A-Z]. Caso seja ascending=False, ordenação decrescente.

# 10 - Funções de Agrupamento (Count(), Sum(), Min(), Max() e Mean())

* No exemplo abaixo, estamos fazendo as seguintes operações:

Falando um pouco do método groupby(), combinamos com o método agg() para realizar o agrupamento e em seguida na coluna valor definimos os seguintes dados:

* Dados Agrupados por:
** QtdBem (Quantidade de Bens por Tipo de Bem) - count()
** VlrTotalBem (Valor Total do Bem por Tipo de Bem) - sum()
** VlrMenorBem (Menor Valor por Tipo de Bem) - min()
** VlrMaiorBem (Maior Valor por Tipo de Bem) - max()
** VlrMedioBem (Média de Valor por Tipo de Bem) - mean()


* Seu funcionamento é bem parecido com o comando GROUP BY de bancos de dados

** Documentação COUNT: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.count.html

** Documentação MIN: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.min.html

** Documentação MAX: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.max.html

# 37.1 - Download de Arquivo de um Arquivo Zipado, Descompactar e Importar

```python
import wget
import os
from zipfile import ZipFile

urlList= ['https://cdn.tse.jus.br/estatistica/sead/odsele/consulta_cand/consulta_cand_2022.zip']
url= 'https://cdn.tse.jus.br/estatistica/sead/odsele/consulta_cand/consulta_cand_2022.zip'
diretorioDownload = '../ProjetoPythonKB/Dados/Zip'
diretorioUnZip = '../ProjetoPythonKB/Dados/Candidato' ## CASO NÃO EXISTE O DIRETÓRIO SERÁ CRIADO

# Download File
for urlFile in urlList:
    
    filename = diretorioDownload + '/' + os.path.basename(urlFile) # get the full path of the file
    
    if os.path.exists(filename):
        
        os.remove(filename) 
        
    wget.download(url, diretorioDownload)
    
diretorioRaiz = filename

# Unzip File    
with ZipFile(diretorioRaiz, 'r') as zipObj:
   
   zipObj.extractall(diretorioUnZip)
```
