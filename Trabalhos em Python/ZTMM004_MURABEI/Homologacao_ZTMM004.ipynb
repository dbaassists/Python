```python

#arquivoSAPBW = 'C:\\Temp\\ZTMM004_CCV\\DA_MM019.20221121.175050'  # 1 Total Registros Arquivo: 28473
#arquivoSAPBW = 'C:\\Temp\\ZTMM004_CCV\\DA_MM019.20221122.160951' # 2 Total Registros Arquivo: 27126
#arquivoSAPBW = 'C:\\Temp\\ZTMM004_CCV\\DA_MM019.20221122.173700' # 3 Total Registros Arquivo: 27124
arquivoSAPBW = 'C:\\Temp\\ZTMM004_CCV\\DA_MM019.20221123.101428' # 4 Total Registros Arquivo: 28472


dfZTMM004_SAPBW = pd.read_csv(arquivoSAPBW, sep=';',usecols=[1,2,3,4,10,11],header=None,skiprows=0
                              ,names=['CodCentro'
                                      ,'CodContaCorrente'
                                      ,'CodMaterial'
                                      ,'CodCliente'
                                      ,'Qtd'
                                      ,'Vlr']
                                 ,dtype={'CodCentro':'str'
                                         ,'CodContaCorrente':'str'
                                         ,'CodMaterial':'str'
                                         ,'CodCliente':'str'
                                         ,'Qtd':'str'
                                         ,'Vlr':'str'})

dfZTMM004_SAPBW['CodCliente'].fillna('0',inplace=True)

dfZTMM004_SAPBW['CodCliente'] = dfZTMM004_SAPBW['CodCliente'].astype(str).str.zfill(10) 

dfZTMM004_SAPBW['Chave'] = dfZTMM004_SAPBW['CodContaCorrente'] \
                                + '_' + dfZTMM004_SAPBW['CodMaterial'] \
                                + '_' + dfZTMM004_SAPBW['CodCentro'] \
                                + '_' + dfZTMM004_SAPBW['CodCliente']



dfZTMM004_SAPBW['ValidaQtd'] = (dfZTMM004_SAPBW['Qtd'].astype(str).str.contains('-').replace('-','')*-1)

dfZTMM004_SAPBW.loc[dfZTMM004_SAPBW['ValidaQtd'] == '-1','ValidaQtd'] = -1
dfZTMM004_SAPBW.loc[dfZTMM004_SAPBW['ValidaQtd'] == 0,'ValidaQtd'] = 1

dfZTMM004_SAPBW['Qtd'] = (dfZTMM004_SAPBW['Qtd'].astype(str).str.replace('-','')).astype(float) * (dfZTMM004_SAPBW['ValidaQtd']).astype(float)

dfZTMM004_SAPBW['ValidaVlr'] = (dfZTMM004_SAPBW['Vlr'].astype(str).str.contains('-').replace('-','')*-1)

dfZTMM004_SAPBW.loc[dfZTMM004_SAPBW['ValidaVlr'] == '-1','ValidaVlr'] = -1
dfZTMM004_SAPBW.loc[dfZTMM004_SAPBW['ValidaVlr'] == 0,'ValidaVlr'] = 1

dfZTMM004_SAPBW['Vlr'] = (dfZTMM004_SAPBW['Vlr'].astype(str).str.replace('-','')).astype(float) * (dfZTMM004_SAPBW['ValidaVlr']).astype(float)

dfZTMM004_SAPBW.info()
```

```python
arquivoSAPECC = 'C:\\Temp\\ZTMM004_CCV\\exportZTMM004_20211031.XLSX'

dfZTMM004_SAPECC = pd.read_excel(arquivoSAPECC
                                 ,usecols=[0,2,4,5,7,8]
                                 ,names=['CodContaCorrente'
                                        ,'CodMaterial'
                                        ,'CodCentro'
                                        ,'CodCliente'
                                        ,'Qtd'
                                        ,'Vlr']
                                 ,dtype={'CodContaCorrente':'str'
                                         ,'CodMaterial':'str'
                                         ,'CodCentro':'str'
                                         ,'CodCliente':'str'
                                         ,'Qtd':'float'
                                         ,'Vlr':'float'})


dfZTMM004_SAPECC['CodCliente'].fillna('0',inplace=True)

dfZTMM004_SAPECC['Qtd'].fillna(0.0,inplace=True)
dfZTMM004_SAPECC['Vlr'].fillna(0.0,inplace=True)

dfZTMM004_SAPECC['CodContaCorrente'] = dfZTMM004_SAPECC['CodContaCorrente'].astype(str).str.zfill(3) 

dfZTMM004_SAPECC['CodCliente'] = dfZTMM004_SAPECC['CodCliente'].astype(str).str.zfill(10) 

dfZTMM004_SAPECC['Chave'] = dfZTMM004_SAPECC['CodContaCorrente'] \
                                + '_' + dfZTMM004_SAPECC['CodMaterial'] \
                                + '_' + dfZTMM004_SAPECC['CodCentro'] \
                                + '_' + dfZTMM004_SAPECC['CodCliente']


dfZTMM004_SAPECC = dfZTMM004_SAPECC.groupby(['CodCentro'
                         ,'CodContaCorrente'
                         ,'CodMaterial'
                         ,'CodCliente'
                         ,'Chave'],as_index=False).agg(Qtd=('Qtd','sum')
                                                       ,Vlr=('Vlr','sum'))

dfZTMM004_SAPECC.info()
```

```python
resultZTMM004_1027 = pd.merge(left=dfZTMM004_SAPECC
                              , right=dfZTMM004_SAPBW
                              , left_on='Chave'
                              , right_on='Chave' 
                              , how='left'
                                    )

resultZTMM004_1027.rename(columns={'Qtd_x':'Qtd_SAPBW'
                                   ,'Qtd_y':'Qtd_SAPECC'
                                   ,'Vlr_x':'Vlr_SAPBW'
                                   ,'Vlr_y':'Vlr_SAPECC'
                                   ,'CodContaCorrente_x':'CodContaCorrente'
                                   ,'CodMaterial_x':'CodMaterial'
                                   ,'CodCentro_x':'CodCentro'
                                   ,'CodCliente_x':'CodCliente'},inplace=True)



# ETAPA DE COMPARAÇÃO COM OS DADOS DO SAP BW E OS DADOS DO SAP ECC - Coluna Quantidade
resultZTMM004_1027.loc[resultZTMM004_1027['Qtd_SAPBW'] == resultZTMM004_1027['Qtd_SAPECC'], 'Comparacao_Qtd'] = 'Igual'
resultZTMM004_1027.loc[resultZTMM004_1027['Qtd_SAPBW'] != resultZTMM004_1027['Qtd_SAPECC'], 'Comparacao_Qtd'] = 'Diferente'

# ETAPA DE COMPARAÇÃO COM OS DADOS DO SAP BW E OS DADOS DO SAP ECC - Coluna Valor
resultZTMM004_1027.loc[resultZTMM004_1027['Vlr_SAPBW'] == resultZTMM004_1027['Vlr_SAPECC'], 'Comparacao_Vlr'] = 'Igual'
resultZTMM004_1027.loc[resultZTMM004_1027['Vlr_SAPBW'] != resultZTMM004_1027['Vlr_SAPECC'], 'Comparacao_Vlr'] = 'Diferente'

resultComparacao_ZTMM004_1027 = resultZTMM004_1027[['Chave'
                                                    ,'CodContaCorrente'
                                                    ,'CodMaterial'
                                                    ,'CodCentro'
                                                    ,'CodCliente'
                                                    ,'Qtd_SAPBW'
                                                    ,'Qtd_SAPECC'
                                                    ,'Comparacao_Qtd'
                                                    ,'Vlr_SAPBW'
                                                    ,'Vlr_SAPECC'
                                                    ,'Comparacao_Vlr']]
```
