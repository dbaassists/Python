```python
import warnings
import pandas as pd
from pandas.core.common import SettingWithCopyWarning
import numpy as np
import json
```

```python
pd.options.display.max_rows = 999
pd.options.display.max_columns=999
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)
```

```python
arquivo = 'C:\\Temp\\ZCON015\\Baixas.XLSX'

dfZCON015_SAP_ECC = pd.read_excel(arquivo
                                  , usecols=[0,2,4,8,9,13,23]
                                  , names=['CodEmpresa'
                                           ,'CodImobilizado'
                                           ,'CodCentro'
                                           ,'NumDocumento'
                                           ,'DtLancamento'
                                           ,'Qtd'
                                          ,'CabDocumento']
                                 , dtype={'CodEmpresa':'str'
                                          ,'CodImobilizado':'str'
                                          ,'CodCentro':'str'
                                          ,'NumDocumento':'str'
                                          ,'DtLancamento':'str'
                                          ,'Qtd':'float64'
                                          ,'CabDocumento':'str'}
                                 , date_parser=[9]
                                 )


dfZCON015_SAP_ECC['CabDocumento'].fillna('0',inplace=True)
dfZCON015_SAP_ECC['CodTipoMovimento'] = dfZCON015_SAP_ECC['CodImobilizado'].str[0:4]
dfZCON015_SAP_ECC['DocZBA'] = dfZCON015_SAP_ECC[dfZCON015_SAP_ECC['CabDocumento'].str.contains("ZBA")]['CabDocumento'].str[0:3]
dfZCON015_SAP_ECC['CodDocumentoZBM'] = dfZCON015_SAP_ECC[dfZCON015_SAP_ECC['CabDocumento'].str.contains("ZBA")]['CabDocumento'].str[4:15]
dfZCON015_SAP_ECC['NumDocumento'] = dfZCON015_SAP_ECC['NumDocumento'].astype(str).str.zfill(10) 
```

```python
dfZCON015_SAP_ECC
```

```python
arquivoBASEZBA = 'C:\\Temp\\ZCON015\\BASE_ZBA.csv'

dfBASEZBA = pd.read_csv(arquivoBASEZBA
                       ,sep=';'
                       ,usecols=[0,1,2]
                       ,names=['CodDocumentoZBM','DocZBA','CodCentro']
                       ,dtype={'CodDocumentoZBM':'str','DocZBA':'str','CodCentro':'str'})

dfBASEZBA['CodDocumentoZBM'] = dfBASEZBA['CodDocumentoZBM'].str[14:30]
```

```python
resultZCON015_SAPECC = pd.merge(left=dfZCON015_SAP_ECC
                              , right=dfBASEZBA
                              , left_on='CodDocumentoZBM'
                              , right_on='CodDocumentoZBM'  
                              , how='left'
                                    )
```

```python
resultZCON015_SAPECC['CodDocumentoZBM'].fillna('0',inplace=True)
resultZCON015_SAPECC[resultZCON015_SAPECC['CodDocumentoZBM'] != '0'].count()
resultZCON015_SAPECC
```

```python
arquivoDAMM020 = 'C:\\Temp\\ZCON015\\DA_MM020.20221107.095828'

dfZCON015_SAP_BW = pd.read_csv(arquivoDAMM020
                               , sep=';'
                               ,header=None
                               ,usecols=[2,4,6,10,11,15,49]
                               ,names=['CodEmpresa'
                                      ,'CodImobilizado'
                                      ,'CodCentro'
                                      ,'NumDocumento'
                                      ,'DtLancamento'
                                      ,'Qtd'
                                      ,'CabDocumento']
                              , dtype={'CodEmpresa':'str'
                                       ,'CodImobilizado':'str'
                                       ,'CodCentro':'str'
                                       ,'NumDocumento':'str'
                                       ,'DtLancamento':'str'
                                       ,'Qtd':'str'
                                       ,'CabDocumento':'str'}
                              , date_parser=[11])

dfZCON015_SAP_BW['CabDocumento'].fillna('0',inplace=True)
dfZCON015_SAP_BW['CodTipoMovimento'] = dfZCON015_SAP_BW['CodImobilizado'].str[0:4]
dfZCON015_SAP_BW['DocZBA'] = dfZCON015_SAP_BW[dfZCON015_SAP_BW['CabDocumento'].str.contains("ZBA")]['CabDocumento'].str[0:3]
dfZCON015_SAP_BW['CodDocumentoZBM'] = dfZCON015_SAP_BW[dfZCON015_SAP_BW['CabDocumento'].str.contains("ZBA")]['CabDocumento'].str[4:15]
dfZCON015_SAP_BW['NumDocumento'] = dfZCON015_SAP_BW['NumDocumento'].astype(str).str.zfill(10) 

#dfZCON015_SAP_BW[dfZCON015_SAP_BW['NumDocumento']=='0010015410']
```

```python
resultZCON015_SAPBW = pd.merge(left=dfZCON015_SAP_BW
                              , right=dfBASEZBA
                              , left_on='CodDocumentoZBM'
                              , right_on='CodDocumentoZBM'  
                              , how='left'
                              )

resultZCON015_SAPBW['CodDocumentoZBM'].fillna('0',inplace=True)

resultZCON015_SAPBW[resultZCON015_SAPBW['CodDocumentoZBM'] != '0'].count()

resultZCON015_SAPBW.head(10)
```

```python
resultZCON015_SAPECC['Chave'] = resultZCON015_SAPECC['CodEmpresa'] + '_' + resultZCON015_SAPECC['CodImobilizado'] + '_' + resultZCON015_SAPECC['CodCentro_x']  + '_' + resultZCON015_SAPECC['NumDocumento']

resultZCON015_SAPECC[resultZCON015_SAPECC['Chave'] == '1001_230000000445_1021_0010006202']
```

```python


resultZCON015_SAPEC_SAPBW = pd.merge(left=resultZCON015_SAPECC
                              , right=resultZCON015_SAPBW
                              , left_on='Chave'
                              , right_on='Chave'  
                              , how='left'
                              )[['Chave'
                                 ,'CodEmpresa_x'
                                 ,'CodImobilizado_x'
                                 ,'CodCentro_x_x'
                                 ,'NumDocumento_x'
                                 ,'DtLancamento_x'
                                 ,'Qtd_x'
                                 ,'Qtd_y'
                                 ,'CabDocumento_x'
                                 ,'CabDocumento_y'
                                 ,'CodDocumentoZBM_x'
                                 ,'CodCentro_y_y']]
```
