```python
arquivoFatoAbastecimentoPRD = 'C:\\Temp\\fato_abastecimento_prd.csv'

dfFatoAbastecimentoPRD = pd.read_csv(arquivoFatoAbastecimentoPRD
                        ,usecols=[0,1,2,3,4]
                        ,names=['CodRelatorio'
                                ,'CodItem'
                                ,'CodDocVenda'
                                ,'NumItemDocVenda'
                                ,'PercNivelTanque']
                        ,dtype={'CodRelatorio' : 'str',
                                'CodItem' : 'str',
                                'CodDocVenda' : 'str',
                                'NumItemDocVenda' : 'str',
                                'PercNivelTanque' : 'float',
                                }
                        ,header=0
                        ,sep=';')

dfFatoAbastecimentoPRD['CodItem'] = pd.Series(dfFatoAbastecimentoPRD['CodItem']).astype(str).str.zfill(6) 
dfFatoAbastecimentoPRD['NumItemDocVenda'] = pd.Series(dfFatoAbastecimentoPRD['NumItemDocVenda']).astype(str).str.zfill(6)
dfFatoAbastecimentoPRD['CodDocVenda'] = pd.Series(dfFatoAbastecimentoPRD['CodDocVenda']).astype(str).str.zfill(10)

dfFatoAbastecimentoPRD['Chave'] = dfFatoAbastecimentoPRD['CodRelatorio'] + '_' + \
                                    dfFatoAbastecimentoPRD['CodItem'] + '_' + \
                                    dfFatoAbastecimentoPRD['NumItemDocVenda'] + '_' + \
                                    dfFatoAbastecimentoPRD['CodDocVenda'] 


#dfFatoAbastecimentoPRD
```

```python
result = pd.merge((pd.merge(left=dfAbastecimentoSergio
                      , right=dfNovaFato
                      , left_on='Chave'
                      , right_on='Chave'
                      , how='left')
                      )
                      , right=dfFatoAbastecimentoPRD
                      , left_on='Chave'
                      , right_on='Chave'
                      , how='left')[['Chave'
                                           ,'CodRelatorio_x'
                                           ,'CodItem_x'
                                           ,'CodDocVenda_x'
                                           ,'NumItemDocVenda_x'
                                           ,'PercNivelTanque_x'
                                           ,'PercNivelTanque_y'
                                           ,'PercNivelTanque']]

result.rename(columns={'CodRelatorio_x':'CodRelatorio'
                          ,'CodItem_x':'CodItem'
                          ,'CodDocVenda_x':'CodDocVenda'
                          ,'NumItemDocVenda_x':'NumItemDocVenda'
                          ,'PercNivelTanque_x':'PercNivelTanque_ARQ_SERGIO'
                          ,'PercNivelTanque_y':'PercNivelTanque_ARQ_NOVA_FATO'
                          ,'PercNivelTanque':'PercNivelTanque_FATO_PRD'},inplace=True)

result['PercNivelTanque_ARQ_SERGIO'].fillna(0,inplace=True)
result['PercNivelTanque_ARQ_NOVA_FATO'].fillna(0,inplace=True)
result['PercNivelTanque_FATO_PRD'].fillna(0,inplace=True)

result.loc[result['PercNivelTanque_ARQ_NOVA_FATO'] == result['PercNivelTanque_FATO_PRD'], 'Comparacao'] = 'Igual'
result.loc[result['PercNivelTanque_ARQ_NOVA_FATO'] != result['PercNivelTanque_FATO_PRD'], 'Comparacao'] = 'Diferente'
```

```python
result.groupby(['Comparacao'],as_index=False).agg(Qtd=('Comparacao','count'))
```
