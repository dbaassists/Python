```python
import warnings
import pandas as pd
from pandas.core.common import SettingWithCopyWarning
import numpy as np
import json
import os
```

```python
pd.options.display.max_rows = 999
pd.options.display.max_columns=999
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)
```

```python
arquivoALEX = 'C:\\Temp\\ZTSD088\\Problemas no extrator.xlsx'

dfALEX = pd.read_excel(arquivoALEX
                       , usecols=(0,1)
                       , names=['CodRelatorio','NumItem']
                            , dtype={'CodRelatorio':'str','NumItem':'str'})

dfALEX.rename(columns={'Número do DCR':'CodRelatorio','Nº do item':'NumItem'}, inplace=True)

dfALEX['NumItem'] = pd.Series(dfALEX['NumItem']).astype(str).str.zfill(6)

dfALEX['CodRelatorio_NumItem'] = dfALEX['CodRelatorio'] + '_' + dfALEX['NumItem']

dfALEX.head(3)
```

```python
arquivoZTSD088 = 'C:\\Temp\\ZTSD088\\VendaDCRLO\\'

dfFinalZTSD088 = pd.DataFrame()

for arquivosZTSD088 in os.listdir(arquivoZTSD088):

    arq_final = arquivoZTSD088 + arquivosZTSD088

    dfZTSD088 = pd.read_csv(arq_final
                            , header=None
                            , sep=';'
                            , low_memory=False
                            , usecols=(0,1)
                            , names=['CodRelatorio','NumItem']
                            , dtype={'CodRelatorio':'str','NumItem':'str'}
                           )
    
    
    dfZTSD088['CodRelatorio_NumItem'] = dfZTSD088['CodRelatorio'] + '_' + dfZTSD088['NumItem']
    
    dfZTSD088['Arquivo']  = arquivosZTSD088
    
    dfFinalZTSD088 = dfFinalZTSD088.append(dfZTSD088)
```

```python
resultZTSD088 = pd.merge(left=dfALEX
                      , right=dfFinalZTSD088
                      , left_on='CodRelatorio_NumItem'
                      , right_on='CodRelatorio_NumItem'
                      )
```
