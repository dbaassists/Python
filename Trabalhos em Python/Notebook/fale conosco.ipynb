```python
import pandas as pd
```

```python
arquivo = 'C:\\Temp\\FAT_FaleConosco - Arquitetura.xlsx'

df = pd.read_excel(arquivo, sheet_name='FAT_CallCenter', engine="openpyxl")

df['DH_InicioTratativa'].replace('00/01/1900 00:00:00','01/01/1900 00:00:00',inplace=True)

df.to_csv('fale_conosco.csv',sep=';',index=False)

df2 = pd.read_csv('fale_conosco.csv',sep=';',parse_dates=['DH_InicioTratativa'],dayfirst=True)

df2['DH_InicioTratativa'] = pd.to_datetime(df2['DH_InicioTratativa'], dayfirst=False, errors = 'coerce')
df2['DH_FimTratativa'] = pd.to_datetime(df2['DH_FimTratativa'], dayfirst=False, errors = 'coerce')
df2['DT_Solicitacao'] = pd.to_datetime(df2['DT_Solicitacao'], dayfirst=False, errors = 'coerce')

df2
```
