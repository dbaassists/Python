```python
paramJson = "C:\\Temp\\paramtanques.json"
print(paramJson)
file = open(paramJson)
dfJsonTanques = json.load(file)
```

```python
def montaCampos(colunas):
        
    vcolunas = [] 
    vcolunasDataType = {}    
    vvalor = 'str'

    for indlistaCampos in colunas.split(sep=','):
        if indlistaCampos != ',':        
            vcolunas.append(indlistaCampos.strip())

    for itemChave in range(len(vcolunas)):
        vcolunasDataType[vcolunas[itemChave]] = vvalor

    return vcolunas, vcolunasDataType
```

```python
colunas = dfJsonTanques['EQUI']['colunas']
vcolunas,vcolunasDataType = montaCampos(colunas)

arquivoEQUI = 'C:\\Temp\\EQUI.csv'

dfEQUI = pd.read_csv(arquivoEQUI
                            ,sep=';'
                            ,skiprows=1
                            ,names=vcolunas
                            ,low_memory=False
                            ,dtype=vcolunasDataType)

#dfEQUI

```

```python
colunas = dfJsonTanques['EQKT']['colunas']
vcolunas,vcolunasDataType = montaCampos(colunas)

arquivoEQKT = 'C:\\Temp\\EQKT.csv'

dfEQKT = pd.read_csv(arquivoEQKT
                            , sep=';'
                            ,skiprows=1
                            ,names=vcolunas
                            ,low_memory=False
                            ,dtype=vcolunasDataType)

#dfEQKT

#dfEquipamento['EQUITYPE'].fillna('Nulo',inplace=True)
```
