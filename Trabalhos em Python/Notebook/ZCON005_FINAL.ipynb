```python
arq_ZCON005_Transacao = 'C:\Temp\exportZCON005_TRANSACAO_20220801.XLSX'

dfZCON005 = pd.read_excel(arq_ZCON005_Transacao
                            , header=None
                            , usecols=(0,1,2,9,10,18)
                            , names=['Empresa'
                                     ,'CodClasseImobilizado'
                                     ,'CodImobilizadoImobilizado'
                                     ,'DtIncorporacao'
                                     ,'Qtd'
                                     ,'CodigoCentro']
                           , dtype={'Empresa':'str'
                                     ,'CodClasseImobilizado':'str'
                                     ,'CodImobilizadoImobilizado':'str'
                                     ,'DtIncorporacao':'str'
                                     ,'Qtd':'float'
                                     ,'CodigoCentro':'str'}
                           )


dfZCON005['Chave'] = dfZCON005['Empresa'] +'_'+ dfZCON005['CodClasseImobilizado'] +'_'+ dfZCON005['CodImobilizadoImobilizado']
```

```python
resultZCON005 = pd.merge(left=dfFinal
                      , right=dfZCON005
                      , left_on='Chave'
                      , right_on='Chave'
                      )

resultZCON005.loc[pd.to_numeric(resultZCON005['Qtd_x']) == resultZCON005['Qtd_y'], 'Comparacao'] = 'Igual'
resultZCON005.loc[pd.to_numeric(resultZCON005['Qtd_x']) != resultZCON005['Qtd_y'], 'Comparacao'] = 'Diferente'
```
